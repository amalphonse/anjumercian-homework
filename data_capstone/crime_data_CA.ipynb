```python
import json
import pandas as pd
with open('/content/drive/MyDrive/Zach_Wilson_Bootcamp/Capstone Project/response_1705618766939.json') as f:
    data = json.load(f)
```

```python
df_read_json = pd.read_json(json.dumps(data), orient='index')
print("DataFrame using pd.read_json() method:")
print(df_read_json)
```

```python
crime_ca_data=df_read_json[0][2]
```

```python
crime_data_df = pd.DataFrame(crime_ca_data)
```

```python
p_table = pd.pivot_table( data=crime_data_df,
                        index=['data_year'],
                        #columns=['Aggravated Assault', 'All Other Offenses (Except Traffic)','Arson', 'Burglary','Curfew and Loitering Law Violations' ],
                        #values='Sales',
                        fill_value="None")
```
